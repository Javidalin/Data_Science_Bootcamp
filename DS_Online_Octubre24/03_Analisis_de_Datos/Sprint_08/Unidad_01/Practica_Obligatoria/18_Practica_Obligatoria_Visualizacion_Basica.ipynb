### #1.2

Recrea la siguiente figura, para ello construye en apartados separados cada grfico y luego haz un apartado en el que se junten los cuatro. Las figuras son:
1. Las ciudades ms pobladas, con los numeros de las poblaciones (tendrs que usar texto y recordar que para centrarlos se necesita poner el argumento `ha` a "center"). (Colores: azul, gris, aceituna, marrn, con trasparencia)
2. Histograma del rea total ocupada por las ciudades. (50 bins, fuente = "Bradley Hand itc", tamao 18, color azul acero)
3. El scatter de rea de agua de un ciudad con su poblacion.
4. El boxplot de la altura de las ciudades.

```python
# Grfico 1: Ciudades ms pobladas
ciudades_mas_pobladas = ciudades_california_limpio.nlargest(5, 'population_total')
colores = ['blue', 'grey', 'olive', 'brown', 'pink']

plt.figure(figsize=(12, 10))

# Grfico 1: Ciudades ms pobladas
plt.subplot(2, 2, 1)
barras = plt.bar(ciudades_mas_pobladas['city'], ciudades_mas_pobladas['population_total'], color=colores, alpha=0.3)
for barra, poblacion in zip(barras, ciudades_mas_pobladas['population_total']):
    plt.text(barra.get_x() + barra.get_width() / 2, barra.get_height(), f'{int(poblacion):,}', ha='center')
plt.title("Ciudades ms pobladas de California")
plt.ylabel("Nmero de habitantes (miles)")

# Grfico 2: Histograma del rea total ocupada por las ciudades
plt.subplot(2, 2, 2)
plt.hist(ciudades_california_limpio['area_total_km2'], bins=50)
plt.title("Histograma del rea Total", fontsize=18,fontname= "Bradley Hand itc", color="steelblue")
plt.xlabel("Superficie (km)")
plt.ylabel("Nmero de Ciudades")

# Grfico 3: Scatter rea de agua vs poblacin
plt.subplot(2, 2, 3)
plt.scatter(ciudades_california_limpio['area_water_km2'], ciudades_california_limpio['population_total'], alpha=0.7, color='green')
plt.title("Poblacin vs rea de Agua")
plt.xlabel("rea de agua (km)")
plt.ylabel("Poblacin Total")

# Grfico 4: Boxplot de la altura
plt.subplot(2, 2, 4)
plt.boxplot(ciudades_california_limpio['elevation_m'], patch_artist=True, boxprops=dict(facecolor='red', color='black'))
plt.title("Boxplot de la Altura de las Ciudades")
plt.ylabel("Altura (m)");
```
